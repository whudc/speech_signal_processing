```python
import pandas as pd

def get_data(loss_file, score_file):
    loss_data = pd.read_csv(loss_file)
    score_data = pd.read_csv(score_file)
    batch = loss_data['Step']
    loss = loss_data['Value']
    score = score_data['Value']
    return batch, loss, score
loss_1_file = "MFCC/run-train_Training Loss_Train-tag-Training Loss (1).csv"
loss_2_file = "FreqCNN/run-train_CNN Training Loss_Train-tag-CNN Training Loss.csv"
score_1_file = "MFCC/run-train_Training Loss_Val-tag-Training Loss (1).csv"
score_2_file = "FreqCNN/run-train_CNN Training Loss_Val-tag-CNN Training Loss.csv"

batch, loss_1, score_1 = get_data(loss_1_file, score_1_file)
_, loss_2, score_2 = get_data(loss_2_file, score_2_file)
length = min(len(loss_1), len(loss_2))
def cut(l, *arrays):
    return [array[:l] for array in arrays]
batch, loss_1, loss_2, score_1, score_2 = cut(length, batch, loss_1, loss_2, score_1, score_2)
```

```python

import matplotlib.pyplot as plt
import numpy as np


#绘制图像
plt.plot(batch,loss_1,'r-')
plt.plot(batch,loss_2,'b-')
#设置图表属性
plt.legend(['MFCC','CNN'])#图例
plt.xlabel('batch')#设置x,y轴标记
plt.ylabel('loss')
plt.title('Loss Comparison')#设置图像标题
#保存图像
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


#绘制图像
plt.plot(batch,score_1,'r-')
plt.plot(batch,score_2,'b-')
#设置图表属性
plt.legend(['MFCC','CNN'])#图例
plt.xlabel('batch')#设置x,y轴标记
plt.ylabel('score')
plt.title('Score Comparison')#设置图像标题
#保存图像
plt.show()
```
